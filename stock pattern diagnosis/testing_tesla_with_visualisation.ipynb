```python
# Kütüphanelerimizi import edelim;
import pandas as pd
import yfinance as yf

from datetime import datetime, timezone
```

```python
# Pandas ayarlamalarını yapalım;
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)
```

```python
# Tesla'ı inceleyeceğimiz hisse senedi olarak seçelim;
tesla = yf.Ticker("TSLA")
```

```python
# Teslanın fiyat datasını kaydedelim;
df = tesla.history(period="10y", start="2023-01-01")
df.reset_index(inplace=True)
df.columns = ["Date", "open", "high", "low", "close", "volume", "Dividends", "Stock Splits"]
df
```

```python
# Original değişken isimlerini bir listede saklayalım;
origin_cols = ["Date", "open", "high", "low", "close", "volume", "Dividends", "Stock Splits"]
```

```python
# tanıma işlemini yapacak olan kodumuzu çağıralım;
from identify_candlestick import recognize_candlestick
```

```python
# Verimizi tanımlanan fonksiyondan geçirelim;
recognize_candlestick(df).head()
```

```python
# Oluşan sütunlara bakalım;
df.columns
```

```python
# Pattern olan sütunları yakalayalım;
col_cdl = [col for col in df.columns if "CDL" in col]
patterns = [col for col in col_cdl if df[col].sum() != 0]
patterns[0:5]
```

```python
# Patterni recognise olan tarihleri, pattern ile birlikte yazdıralım;
for col in patterns:
    print("####################")
    print(col)
    print("####################")
    display(df[df[col] != 0][origin_cols])
```

    Burada ilgili tarihlerin mum grafiğini çizdireceğiz ve html olarak saklayacağız.

```python
from plotly.offline import plot
import plotly.graph_objs as go
```

```python
o = df['open'].astype(float)
h = df['high'].astype(float)
l = df['low'].astype(float)
c = df['close'].astype(float)

trace = go.Candlestick(
            open=o,
            high=h,
            low=l,
            close=c)
data = [trace]
```

```python
plot(data, filename='go_candle_tesla.html')
```
