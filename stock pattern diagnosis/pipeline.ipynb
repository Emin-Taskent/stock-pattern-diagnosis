```python
import pandas as pd
import yfinance as yf

from datetime import datetime, timezone
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 500)
```

```python
from identify_candlestick import recognize_candlestick
```

```python
def stock_search(ticker, strength = 3):
    ticker_x = yf.Ticker(ticker)
    dataframe = ticker_x.history(period="1mo")
    dataframe.reset_index(inplace=True)
    dataframe.columns = ["Date", "open", "high", "low", "close", "volume", "Dividends", "Stock Splits"]
    origin_cols = ["Date", "open", "high", "low", "close", "volume", "Dividends", "Stock Splits"]
    recognize_candlestick(dataframe)
    unknown= ['CDL2CROWS',
              'CDL3STARSINSOUTH',
              'CDLADVANCEBLOCK',
              'CDLBELTHOLD',
              'CDLBREAKAWAY',
              'CDLCLOSINGMARUBOZU',
              'CDLCONCEALBABYSWALL',
              'CDLGAPSIDESIDEWHITE',
              'CDLHIGHWAVE',
              'CDLHIKKAKE',
              'CDLHIKKAKEMOD',
              'CDLHOMINGPIGEON',
              'CDLIDENTICAL3CROWS',
              'CDLINNECK',               
              'CDLLADDERBOTTOM',
              'CDLLONGLEGGEDDOJI',                           
              'CDLMATCHINGLOW',
              'CDLMATHOLD',                                 
              'CDLONNECK',                            
              'CDLRICKSHAWMAN',
              'CDLSEPARATINGLINES',
              'CDLSTICKSANDWICH',
              'CDLTAKURI',
              'CDLTASUKIGAP',
              'CDLTHRUSTING',
              'CDLTRISTAR',
              'CDLUNIQUE3RIVER',
              'CDLUPSIDEGAP2CROWS',
              'CDLXSIDEGAP3METHODS']
    
    two_candle = ['CDLSHOOTINGSTAR',
                  'CDLSPINNINGTOP',
                  'CDLPIERCING',
                  'CDLMARUBOZU',
                  'CDLINVERTEDHAMMER',
                  'CDLKICKING',
                  'CDLGRAVESTONEDOJI',
                  'CDLHAMMER',
                  'CDLHANGINGMAN',
                  'CDLHARAMI',
                  'CDLHARAMICROSS',
                  'CDLDARKCLOUDCOVER',
                  'CDLDOJI',
                  'CDLDOJISTAR',
                  'CDLDRAGONFLYDOJI',
                  'CDLENGULFING']
    
    three_candle = ['CDLABANDONEDBABY',
                    'CDLMORNINGDOJISTAR',
                    'CDLMORNINGSTAR',
                    'CDL3WHITESOLDIERS',
                    'CDL3LINESTRIKE',
                    'CDL3OUTSIDE',
                    'CDL3INSIDE',
                    'CDL3BLACKCROWS',
                    'CDLEVENINGDOJISTAR',
                    'CDLEVENINGSTAR',
                    'CDLRISEFALL3METHODS',]
    
    if strength == 3:
        patterns = [col for col in three_candle if dataframe[col].sum() != 0]
        for col in patterns:
            print(f"~~~~~~~~~~~~~~~~~~{ticker}~~~~~~~~~~~~~~~~~~")
            print("####################")
            print(col)
            print("####################")
            display(dataframe[dataframe[col] != 0][origin_cols])
    elif strength == 2:
        patterns = [col for col in two_candle if dataframe[col].sum() != 0]
        for col in patterns:
            print(f"~~~~~~~~~~~~~~~~~~{ticker}~~~~~~~~~~~~~~~~~~")
            print("####################")
            print(col)
            print("####################")
            display(dataframe[dataframe[col] != 0][origin_cols])
    elif strength == 1:
        patterns = [col for col in unknown if dataframe[col].sum() != 0]
        print(f"~~~~~~~~~~~~~~~~~~{ticker}~~~~~~~~~~~~~~~~~~")
        for col in patterns:
            print(f"~~~~~~~~~~~~~~~~~~{ticker}~~~~~~~~~~~~~~~~~~")
            print("####################")
            print(col)
            print("####################")
            display(dataframe[dataframe[col] != 0][origin_cols])


```

```python
sym_df = pd.read_csv("constituents.csv")
sym = [i for i in sym_df["Symbol"]]
```

```python
for i in sym[20:40]:
    stock_search(i, strength=3)
```
